```python
import numpy as np
import pandas as pd
import re
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
driver = './chromedriver.exe'
```

```python
browser = webdriver.Chrome(driver)
```

```python
browser.get('https://www.marketsandmarkets.com/telecom-and-IT-market-research-113.html')
```

```python
cards = browser.find_elements_by_tag_name('tr')
```

```python
len(cards)
```

```python
report_title = []
date_published = []
report_description = []
link = []
```

```python
count = 0
for card in cards[1:50]:
    count += 1
    card_html = BeautifulSoup(card.get_attribute('innerHTML'), 'html.parser')
    title_tag = card_html.find('h3')
    if title_tag == None:
        print(count)       
    else:
        title = title_tag.text
        report_title.append(title)
    
    link_tag = card_html.find('a')
    link_add = link_tag['href']
    link.append('https://www.marketsandmarkets.com' + link_add)
    
    date_tag = card_html.find('td',class_ = 'displaynone')
    date = date_tag.text.replace('\xa0',' ').strip()
    date_published.append(date)
    
    des_tag = card_html.find('p')
    des = des_tag.text
    report_description.append(des)
```

```python
df = pd.DataFrame({'Report Title': report_title,
                  'Date Published': date_published,
                  'Link': link,
                  'Report Description': report_description})
```

```python
per = []
usd = []
```

```python
def extract_percent(des):
    try:
        x = des.find('%')
        if des[x-1] == ' ':
            d = des[:x-1] + des[x:]
        else:
            d = des
        tup = re.findall('\s(100|(\d{1,2}(\.\d+)*))%', d) 
        res = tup[0][0]
    except:
        res = np.nan
    return res
```

```python
def extract_val(des):
    lt = des.split()
    try:
        x = lt.index('USD')
        if lt[x+1][1].isalpha()==False:
            val = lt[x:x+3]
        else:
            new = lt[x+2:]
            y = new.index('USD')
            val = new[y:y+3]
        s_val = ' '.join([elem for elem in val])
    except:
        s_val = np.nan
    return s_val
```

```python
df['USD_value'] = df['Report Description'].apply(extract_val)
```

```python
df['%_val'] = df['Report Description'].apply(extract_percent)
```
